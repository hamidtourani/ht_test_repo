```python
import pandas as pd
```

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Eve', 'Bob'],
        'Age': [25, '30', 28, 25, '22', None, 30],
        'City': ['  New York', 'London ', 'paris', 'new York', ' london', '  ', 'london'],
        'Salary': [60000, 70000, 65000, 60000, 55000, 75000, 70000],
        'Date': ['2023-01-15', '2023-02-20', '2023/03/10', '2023-01-15', '2023-04-05', 'invalid', '2023-02-20']}

mydf = pd.DataFrame(data)
##
```

```python
mydf
```

```python
mydf.isnull().sum()
```

```python
mydf['Age'] = pd.to_numeric(mydf['Age'] , errors='coerce')
mydf['Date'] = pd.to_datetime(mydf['Date'], format='mixed', errors='coerce')
mydf
```

```python
mydf['Age'] = mydf['Age'].fillna(round(mydf['Age'].mean()))
mydf
```

```python
mydf['Date'] = mydf['Date'].fillna(mydf['Date'].min())
mydf
```

```python
mydf["City"] = mydf["City"].str.strip().str.title()
mydf
```

```python
mydf = mydf[mydf["City"].notna() & (mydf["City"] != "")]
```
